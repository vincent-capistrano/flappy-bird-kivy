**Run each cell in order.** At the end you will download a `.apk` file ready to install on Android.

```python
# ── Step 1: System dependencies ───────────────────────────────
!sudo apt-get update -qq
!sudo apt-get install -y -qq \
    git zip unzip openjdk-17-jdk python3-pip \
    autoconf libtool pkg-config zlib1g-dev \
    libncurses5-dev libncursesw5-dev libtinfo5 \
    cmake libffi-dev libssl-dev
print('System packages installed.')
```

```python
# ── Step 2: Python build tools ────────────────────────────────
!pip install -q buildozer cython==0.29.37
print('buildozer + cython installed.')
```

```python
# ── Step 3: Upload your project files ─────────────────────────
# A file picker will appear — upload BOTH:
#   main.py
#   buildozer.spec
from google.colab import files
print('Select main.py AND buildozer.spec to upload:')
uploaded = files.upload()
for name in uploaded:
    print(f'  Uploaded: {name} ({len(uploaded[name])} bytes)')
```

```python
# ── Step 4: Build the debug APK ───────────────────────────────
# This takes 10-20 minutes the first time (downloads Android SDK/NDK).
# Subsequent builds are much faster.
!buildozer android debug 2>&1
```

```python
# ── Step 5: Download the APK ──────────────────────────────────
import glob
from google.colab import files

apks = glob.glob('bin/*.apk')
if apks:
    apk_path = apks[0]
    print(f'Downloading: {apk_path}')
    files.download(apk_path)
else:
    print('No APK found — scroll up to check the build output for errors.')
```
